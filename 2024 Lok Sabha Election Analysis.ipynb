# Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
data = pd.read_csv('election_results_2024.csv')
```

```python
data.head(10)
```

# Party with highest ans lowest margins of victory

```python
data['Margin'] = pd.to_numeric(data['Margin'], errors='coerce')

# Group by Leading Party and calculate total margin of victory
party_margin = data.groupby('Leading Party')['Margin'].sum()

# Find party with highest and lowest margin of victory
if not party_margin.empty:
    highest_margin_party = party_margin.idxmax()
    lowest_margin_party = party_margin.idxmin()
    
    # Print results
    print(f"Party with highest margin of victory: {highest_margin_party}")
    print(f"Party with lowest margin of victory: {lowest_margin_party}")
else:
    print("No data available.")








```

# Plot the number of seats won by each party

```python
# Calculate party votes (assuming you've already grouped and sorted as 'party_votes')
party_votes = data.groupby('Leading Party')['Margin'].sum().sort_values(ascending=False)

# Find party with highest and lowest votes
leading_party_highest_votes = party_votes.idxmax()
leading_party_lowest_votes = party_votes.idxmin()

# Number of seats won by each party
seats_won = data['Leading Party'].value_counts()

# Plot number of seats won by each party
plt.figure(figsize=(20, 8))
sns.barplot(x=seats_won.index, y=seats_won.values, palette='bright')
plt.title('Number of Seats Won by Each Party')
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.xticks(rotation=90)
plt.show()

# Display results
print(f"Party with highest votes: {leading_party_highest_votes}")
print(f"Party with lowest votes: {leading_party_lowest_votes}")
```

# Get the votes for Rahul Gandhi, Narendra Modi, and Amit Shah


```python
# Filter entries for each candidate
rahul_entries = data[data['Leading Candidate'] == 'RAHUL GANDHI']
modi_entries = data[data['Leading Candidate'] == 'NARENDRA MODI']
amit_entries = data[data['Leading Candidate'] == 'AMIT SHAH']

# Get the votes for each candidate
rahul_votes = rahul_entries['Margin'].values
modi_votes = modi_entries['Margin'].values[0] if not modi_entries.empty else 0
amit_votes = amit_entries['Margin'].values[0] if not amit_entries.empty else 0

# Get the original constituency names for each candidate
rahul_constituencies = list(rahul_entries['Constituency'])
modi_constituency = modi_entries['Constituency'].values[0] if not modi_entries.empty else "Modi's Constituency"
amit_constituency = amit_entries['Constituency'].values[0] if not amit_entries.empty else "Amit's Constituency"

# Combine the data into a DataFrame for plotting
data_to_plot = pd.DataFrame({
    'Candidate': ['Rahul Gandhi'] * len(rahul_votes) + ['Narendra Modi', 'Amit Shah'],
    'Constituency': rahul_constituencies + [modi_constituency, amit_constituency],
    'Votes': list(rahul_votes) + [modi_votes, amit_votes]
})

# Plot the comparison
plt.figure(figsize=(12, 6))
sns.barplot(data=data_to_plot, x='Constituency', y='Votes', hue='Candidate', palette='dark')
plt.title('Comparison of Votes for Rahul Gandhi, Narendra Modi, and Amit Shah')
plt.xlabel('Constituency')
plt.ylabel('Votes')
plt.xticks(rotation=45)
plt.show()
```

# Highest and lowest Victory Candidate


```python
# Find entries with highest and lowest margin of victory
highest_margin_entry = data.loc[data['Margin'].idxmax()]
lowest_margin_entry = data.loc[data['Margin'].idxmin()]

# Combine the data into a DataFrame for plotting
data_to_plot = pd.DataFrame({
    'Candidate': [highest_margin_entry['Leading Candidate'], lowest_margin_entry['Leading Candidate']],
    'Party': [highest_margin_entry['Leading Party'], lowest_margin_entry['Leading Party']],
    'Margin': [highest_margin_entry['Margin'], lowest_margin_entry['Margin']]
})

# Plot the comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=data_to_plot, x='Candidate', y='Margin', hue='Party', palette='dark')
plt.title('Candidates with Highest and Lowest Margin of Victory')
plt.xlabel('Candidate')
plt.ylabel('Margin of Victory')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot the cumulative distribution of Margin of Victory
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=data['Margin'])
plt.title('Cumulative Distribution of Margin of Victory')
plt.xlabel('Margin of Victory')
plt.ylabel('Cumulative Proportion')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data['Margin'], fill=True)
plt.title('Kernel Density Estimate of Margin of Victory')
plt.xlabel('Margin of Victory')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data['Margin'])
plt.title('Box Plot of Margin of Victory')
plt.xlabel('Margin of Victory')
plt.ylabel('Distribution')
plt.grid(True)
plt.show()
```

# Votes distribution by party


```python
party_votes = data.groupby('Leading Party')['Margin'].sum().sort_values(ascending=False)
# Plot pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(party_votes, labels=None, autopct='%1.1f%%', startangle=140, wedgeprops=dict(edgecolor='w'))
plt.title('Votes Distribution by Party', pad=20)
plt.axis('equal')
plt.legend(labels=party_votes.index, loc='center left', bbox_to_anchor=(1, 0.5), fontsize='medium')
plt.show()

```

# Top 10 trailing party by VOTE

```python
trailing_party_votes = data.groupby('Trailing Party')['Margin'].sum().sort_values(ascending=False)
trailing_party_seats = data['Trailing Party'].value_counts()
```

```python
plt.figure(figsize=(20, 6))
# Plot votes distribution by trailing party
plt.subplot(1, 2, 1)
sns.barplot(x=trailing_party_votes.index[:10], y=trailing_party_votes.values[:10], palette='colorblind')
plt.title('Top 10 Trailing Parties by Votes')
plt.xlabel('Party')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
```
